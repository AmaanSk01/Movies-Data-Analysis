```python
movies_data = pd.read_csv("Movies.csv")
```

```python
# Step 2: Dropping unnessaryy columns
```

```python
print(movies_data.columns)
```

```python
movies_data.drop(columns=['color', 'director_facebook_likes', 'actor_3_facebook_likes', 
                   'actor_2_facebook_likes', 'actor_1_facebook_likes', 
                   'movie_facebook_likes', 'aspect_ratio', 'content_rating'], inplace=True)
```

```python
print(movies_data.columns)
```

```python
# Step 3: Handling missing values
# Drop rows where crucial columns like 'director_name', 'duration', or 'genres' are missing
movies_data_cleaned = movies_data.dropna(subset=['director_name', 'duration', 'genres'])
```

```python
movies_data_cleaned.loc[:, 'budget'] = movies_data_cleaned['budget'].fillna(movies_data_cleaned['budget'].median())
movies_data_cleaned.loc[:, 'gross'] = movies_data_cleaned['gross'].fillna(movies_data_cleaned['gross'].median())
```

```python
movies_data.head()
```

```python
# Step 4: Removing duplicates
movies_data_cleaned = movies_data_cleaned.drop_duplicates()
```

```python
movies_data.head()
```

```python
# Step 5: Data type conversion
# Convert budget and gross to integer type
movies_data_cleaned['budget'] = movies_data_cleaned['budget'].astype(int)
movies_data_cleaned['gross'] = movies_data_cleaned['gross'].astype(int)
```

```python
movies_data.head()
```

```python
# Step 6: Renaming columns for better readability
movies_data_cleaned.rename(columns={
    'num_critic_for_reviews': 'critic_reviews', 
    'num_user_for_reviews': 'user_reviews', 
    'title_year': 'year',
    'imdb_score': 'imdb_rating'
}, inplace=True)
```

```python
movies_data.head()
```

```python
# Step 7: Detect and handle outliers (capping extreme values in 'gross' and 'budget')
movies_data_cleaned['budget'] = movies_data_cleaned['budget'].clip(
    lower=movies_data_cleaned['budget'].quantile(0.05),
    upper=movies_data_cleaned['budget'].quantile(0.95)
).infer_objects(copy=False)

# Clip the 'gross' column and ensure the correct type
movies_data_cleaned['gross'] = movies_data_cleaned['gross'].clip(
    lower=movies_data_cleaned['gross'].quantile(0.05),
    upper=movies_data_cleaned['gross'].quantile(0.95)
).infer_objects(copy=False)
```
